```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the dataset
file_path = 'wbc.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

## Langkah 1: Pisahkan Variabel yang Dapat Digunakan dan Tidak Dapat Digunakan

Berdasarkan tampilan data awal:

-  Kolom 'id': Tidak relevan untuk analisis diagnostik, sehingga akan dihapus.
-  Kolom 'Unnamed: 32': Mengandung nilai NaN dan tampaknya tidak relevan, sehingga akan dihapus.



```python
# Remove columns that are not useful for the analysis
data_cleaned = data.drop(columns=['id', 'Unnamed: 32'])

# Display the cleaned dataset
data_cleaned.head()
```

## Encoding

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
data_cleaned['diagnosis'] = le.fit_transform(data['diagnosis'])
```

```python
data_cleaned.head()
```

## Standarisasi

```python
std = StandardScaler()
numeric_columns = data_cleaned.columns[data_cleaned.columns != 'diagnosis'] # mengecualikan kolom diagnosis karena sudah di encode diatas
data_cleaned[numeric_columns] = std.fit_transform(data_cleaned[numeric_columns])
data_cleaned
```

## Stratified Split Data Rasio 80:20

```python
from sklearn.model_selection import train_test_split
# Step 4: Perform stratified split into training and testing datasets


df2_train, df2_test = train_test_split(data_cleaned, test_size=0.2, random_state=0, stratify=data_cleaned['diagnosis'])


# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{data_cleaned.shape[0]}')
print(f'Jumlah label data train:\n{df2_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df2_test.diagnosis.value_counts()}')
```
